```python
import pandas as pd
```

```python
df = pd.read_csv("customer_segmentation.csv")
df
```

```python
df = df.drop("customer_id", axis=1)
df
```

```python
df.info()
```

```python
df.describe(include = 'all')
```

```python
mask_average_price_positive_or_null = (df.average_price > 0) | (df.average_price.isnull())
df = df[mask_average_price_positive_or_null]
df
```

```python
df.describe(include = 'all')
df
```

```python
df.isnull().sum() / df.shape[0]
```

```python
df = df.drop('hobbies', axis=1)
df
```

```python
gender_mode = df["gender"].mode()[0]
df.loc[:, "gender"] = df["gender"].fillna(gender_mode)
```

```python
means = df[["average_price", "age"]].mean()
df.loc[:, ["average_price", "age"]] = df[["average_price", "age"]].fillna(means)
```

```python
df.info()
```

```python
y = df["client_type"]
X = df.drop("client_type", axis=1)
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()
y = labelencoder.fit_transform(y)

print(y)
print(labelencoder.classes_)
```

```python
from sklearn.preprocessing import OneHotEncoder

onehotencoder = OneHotEncoder(sparse=False)
X_dummy = onehotencoder.fit_transform(X[["country", "gender"]])

print(X_dummy)
print(onehotencoder.categories_)
```

```python
import numpy as np
dummy_cols = np.concatenate((onehotencoder.categories_[0], onehotencoder.categories_[1]), axis=None)
print(dummy_cols)
```

```python
onehotencoder.get_feature_names_out()
```

```python
X_dummy = pd.DataFrame(X_dummy, columns=onehotencoder.get_feature_names_out())
X_dummy
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_num = scaler.fit_transform(X[["average_basket", "average_price", "visit_number", "age"]])
```

```python
X_num
```

```python
X_num = pd.DataFrame(X_num, columns=["average_basket", "average_price", "visit_number", "age"])
```

```python
X_num
```

```python
X = pd.concat([X_dummy, X_num], axis=1)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
my_tree_classifier = DecisionTreeClassifier()
```

```python
my_tree_classifier.fit(X_train, y_train)
```

https://scikit-learn.org/stable/tutorial/machine_learning_map/index.html

```python
my_tree_classifier.score(X_train, y_train)
```

```python
my_tree_classifier.score(X_test, y_test)
```

```python
df_new = pd.read_csv("customer_segmentation_new.csv")
df_new = df_new.drop("customer_id", axis=1)
mask_average_price_positive_or_null = (df_new.average_price > 0) | (df_new.average_price.isnull())
df_new = df_new[mask_average_price_positive_or_null]
df_new = df_new.drop("hobbies", axis=1)
df_new.loc[:, "gender"] = df_new["gender"].fillna(gender_mode)
df_new.loc[:, ["average_price", "age"]] = df_new[["average_price", "age"]].fillna(means)

X_new = df_new
# Encode categorical features + Scale numerical features 
X_new_dummy = pd.DataFrame(onehotencoder.transform(X_new[["country", "gender"]]), columns=onehotencoder.get_feature_names_out())
X_new_num = scaler.transform(X_new[["average_basket", "average_price", "visit_number", "age"]])
X_new_num = pd.DataFrame(X_new_num, columns=["average_basket", "average_price", "visit_number", "age"])
X_new = pd.concat([X_new_dummy, X_new_num], axis=1)

# Predict
y_predict = my_tree_classifier.predict(X_new)
```
